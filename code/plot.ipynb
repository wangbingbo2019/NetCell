```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
 

```

```python
#读取输入文件
insumstats = pd.read_csv(r"E:\00_CeSOP\plot\MegaGWAS_summary_Asian_ChrX.txt.gz","\\s+")
```

```python
insumstats["log-pvalue"] = -np.log10(insumstats["PVALUE"])
```

```python
import gwaslab as gl
```

```python
# 3. 设置曼哈顿图的背景样式
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

# 4. 绘制曼哈顿图
# 循环每个染色体，按染色体位置画出每个SNP的log10(P-value)
for chrom in insumstats['CHR'].unique():
    chrom_data = insumstats[insumstats['CHR'] == chrom]
    plt.scatter(chrom_data['POS'], chrom_data['log-pvalue'], s=10, label=f"Chromosome {chrom}")

# 5. 设置x轴和y轴标签
plt.xlabel('染色体', fontsize=12)
plt.ylabel('-log10(P-value)', fontsize=12)
plt.title('GWAS', fontsize=16)

# 6. 添加显著性阈值线（通常是P-value=0.05的-Log10(P-value)=1.3）
plt.axhline(y=-np.log10(0.05), color='red', linestyle='--', label='Significance threshold')

# 7. 设置x轴标签的旋转角度
plt.xticks(rotation=90)

# 8. 优化显示
plt.tight_layout()

# 9. 显示图形
plt.show()
```
